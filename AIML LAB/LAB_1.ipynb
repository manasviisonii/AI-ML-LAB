```python
missing_values_per_column=df.isnull().sum()
print(missing_values_per_column)
```

```python
total_missing_values=missing_values_per_column.sum()
print(total_missing_values)
```

```python
rows_with_missing_values=df.isnull().any(axis=1)
num_rows_with_missing_values=rows_with_missing_values.sum()
print(rows_with_missing_values)
```

```python
df.describe().max()
```

```python
mean=df.select_dtypes(include=['number']).mean()
print(mean)
```

```python
df.fillna(mean,inplace=True)
```

```python
min=df.select_dtypes(include=['number']).min()
print(min)
```

```python
max=df.select_dtypes(include=['number']).max()
print(max)
```

```python
sum=df.select_dtypes(include=['number']).sum()
print(sum)
```

```python
df.to_csv('train_save_1.csv',index=False)
```
